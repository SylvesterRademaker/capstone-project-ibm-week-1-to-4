## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp as ConvertedComp
FROM master

"""
df_Converted = pd.read_sql_query(QUERY,conn)
df_Converted.plot.hist( color='orange', edgecolor='black')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT Age as Age
FROM master

"""
df_Age = pd.read_sql_query(QUERY,conn)
df_Age.plot.box()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import matplotlib.pyplot as plt 
```

```python
# your code goes here

QUERY2 = """
SELECT WorkWeekHrs as WorkWeekHrs
FROM master

"""
df_WorkWeekHrs = pd.read_sql_query(QUERY2,conn)
x = df_WorkWeekHrs
y = df_Age
plt.figure(figsize=(12, 8))
plt.scatter(x, y, color='blue', marker='o', s=50)
plt.xlabel('Work week hours')
plt.ylabel('Age')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY3 = """
SELECT CodeRevHrs as CodeRevHrs
FROM master

"""
df_CodeRevHrs = pd.read_sql_query(QUERY3,conn)
df_Age_cleaned = df_Age.dropna()
df_CodeRevHrs_cleaned = df_CodeRevHrs.dropna()
df_WorkWeekHrs_cleaned = df_WorkWeekHrs.dropna()
import plotly.graph_objects as go
hover_text = [f"Work hours per week: {work_hours}<br>Code Review Hours: {code_rev_hours}<br>Age: {age}" 
              for work_hours, code_rev_hours, age in zip(df_WorkWeekHrs['WorkWeekHrs'], 
                                                          df_CodeRevHrs['CodeRevHrs'], 
                                                          df_Age_cleaned)]

fig = go.Figure(data=[go.Scatter(
    x=df_WorkWeekHrs['WorkWeekHrs'],  
    y=df_CodeRevHrs['CodeRevHrs'],
    mode='markers',
    marker=dict(size=df_Age_cleaned),
    hovertext=hover_text 
)])
fig.update_layout(
    xaxis_title='Work hours per week',
    yaxis_title='Code Review Hours'
)
fig.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import plotly.express as px

QUERY5 = """
SELECT DatabaseDesireNextYear, COUNT(*) AS Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
df_top_databases = pd.read_sql_query(QUERY5, conn)

fig = px.pie(df_top_databases, values='Count', names='DatabaseDesireNextYear', title='Top 5 Databases Respondents Wish to Learn Next Year')

fig.update_traces(textposition='inside', textinfo='percent+label')

fig.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import matplotlib.pyplot as plt

QUERY6 = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df_ag_30t35 = pd.read_sql_query(QUERY6, conn)

x = df_ag_30t35['Age']
y1 = df_ag_30t35['WorkWeekHrs'].median()
y2 = df_ag_30t35['CodeRevHrs']
 
fig, ax1 = plt.subplots()

ax1.bar(x, y1, color='grey', label='Work Week Hours')
ax1.set_xlabel('Age')
ax1.set_ylabel('Work Week Hours')

ax2 = ax1.twinx()
ax2.bar(x, y2, bottom=y1, color='black', label='Code Review Hours')
ax2.set_ylabel('Code Review Hours')

ax1.set_ylim([0, 50])
ax2.set_ylim([0, 60])

plt.title('Work Week vs Code Review Hours')

fig.legend(loc='upper right', bbox_to_anchor=(1.3, 0.9))
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import matplotlib.pyplot as plt

QUERY8 = """
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60
"""

df_ccomp_45t60 = pd.read_sql_query(QUERY8, conn)

median_comp_by_age = df_ccomp_45t60.groupby('Age')['ConvertedComp'].median()

median_comp_by_age.plot(kind='line')

plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages between 45 and 60')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY9 = """
SELECT MainBranch
FROM master
"""

df_MainBranch = pd.read_sql_query(QUERY9, conn)

branch_counts = df_MainBranch['MainBranch'].value_counts()
branch_counts.replace({"I am not primarily a developer, but I write code sometimes as part of my work": "i am not primarily a developer"}, inplace=True)


branch_counts.plot(kind='barh')

plt.xlabel('Count')
plt.ylabel('Main Branch') 
plt.title('Distribution of Main Branch')


plt.show()

```
